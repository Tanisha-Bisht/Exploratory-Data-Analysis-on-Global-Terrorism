```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#fetching data from dataset
data=pd.read_csv("Global Terrorism - START data/globalterrorismdb_0718dist.csv",encoding="latin",low_memory=False)
data.head()
```

```python
data.describe()
```

```python
data.dtypes
```

<b>Removing Duplicacy</b>

```python
data.dropna(thresh=180000,axis=1,inplace=True)
```

```python
data.corr()
```

## <b>Univariate analysis</b>

```python
plt.figure(figsize=(20,10))
sb.countplot(x='iyear',hue='success',data=data)
plt.title('Trend of increase in Terror from 1970 to 2017')
plt.xlabel('Attack Year')
plt.ylabel('Number of attacks')
```

#### <u><b>Note: As we can se that there is drastic increase in number of attacks after 2004</u><b>

```python
data['region_txt'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sb.countplot(x='region_txt',data=data)
plt.xticks(rotation=70)
plt.title('Total number of attacks Region-wise')
plt.xlabel('Region-wise')
plt.ylabel('Number of attacks')
plt.show()
```

#### <b>Region-wise Middle East& NorthAfrica ,South Asia are hot terror zones where as Central Asia has least number of terror attack</b>

```python
data['country_txt'].value_counts()[:20]
```

```python
plt.figure(figsize=(15,6))
sb.barplot(x=data['country_txt'].value_counts()[:20].index,y=data['country_txt'].value_counts()[:20].values)
plt.xticks(rotation=70)
plt.title('Total number of attacks Country-wise')
plt.xlabel('Country')
plt.ylabel('Total attack counts')
plt.show()
```

#### <b>As we can clearly see that Iraq has the higgest number of terror attack and Egypt has lower</b>

```python
data['gname'].value_counts()[:20]
```

```python
plt.figure(figsize=(15,6))
sb.barplot(x=data['gname'].value_counts()[:20].index,y=data['gname'].value_counts()[:20].values)
plt.xticks(rotation=90)
plt.title('Total number of attacks group-wise')
plt.xlabel('Country')
plt.ylabel('Total attack counts')
plt.show()
```

#### <b>From the above data we can see that some unknown groups and Talian,ISIL etc are the major terrorist groups</b>

```python
data['attacktype1_txt'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sb.barplot(x=data['attacktype1_txt'].value_counts().index,y=data['attacktype1_txt'].value_counts().values)
plt.xticks(rotation=70)
plt.title('Total count of Attack types')
plt.xlabel('Attack type')
plt.ylabel('Total attack counts')
plt.show()
```

#### <b>From the above data the major destructive attacktypes are Bombing/Explosion, Armed assault ,Assassination amd many more</b>

```python
data['targtype1_txt'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sb.barplot(x=data['targtype1_txt'].value_counts().index,y=data['targtype1_txt'].value_counts().values)
plt.xticks(rotation=70)
plt.title('Total count of attack on differnt target type')
plt.xlabel('Target type')
plt.ylabel('Total attack counts')
plt.show()
```

#### <b>Attackers target the people who are private citizens and have property ,Military person,police ,Government(General) and many more</b>
